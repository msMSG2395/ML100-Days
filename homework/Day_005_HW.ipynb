```python
app_train.select_dtypes(include=["float64","int64"]) # 顯示數值型變量
```

```python
app_train.describe(include='all') # 進行描述性統計
```

```python
# 對數值型變量進行描述性統計
# 有count(非空值數)、mean(算術平均數)、std(標準差)、min(最小值)、max(最大值)
# 25%(落在下四分位數的值的總計)、50%(落在中位數的值的總計)、75%(落在上四分位數的值的總計)
at = app_train.describe(include=np.number).T.drop(['25%','50%','75%'],axis=1) \
    .assign(missing_pct=app_train.apply(lambda x: (len(x)-x.count())/len(x)),
           nunique = app_train.apply(lambda x: x.nunique()))
```

```python
at[at['missing_pct']==0]
at
```

```python
# 對離散型變量進行描述性統計
# 字段有count（非空值數）、unique（唯一值數）、top（頻數最高者）、freq（最高頻數）
app_train.describe(include=np.object).T
```
